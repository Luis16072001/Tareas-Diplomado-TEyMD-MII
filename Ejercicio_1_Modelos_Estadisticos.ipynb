## Modelo de probabilidad Laplaciano (cl치sico):

**Ejemplo** Se lanza una moneda 3 veces, de manera independiente (es decir, el lanzamienro $i$-칠simo no afecta al siguiente $i+1$).
Tenemos que el espacio muestral de este experiemto es:

$$ \Omega = \{A,S\}칑\{A,S\}칑\{A,S\} = \{A,S\}^{3}$$
donde
$$\#\Omega = 2^{3} = 8.$$

Sea $E$ un evento, entonces la probabilidad de $E$ se calcula como sigue:
$$\mathbb{P}(E) = \frac{\#E}{\#\Omega}$$


```python
# Para crear a Omega (el espacio muestral), necesitamos una librer칤a

import itertools #importamos el m칩dulo
from itertools import product

Omega = set(product({"A","S"},repeat=3))
Omega
```

Consideremos los siguientes eventos:
1. A: todos los resultado son 치guila
2. S: todos los resultados son soles
3. E: Por lo menos un resultado es sol  

Sea el evento $A_{i}$ donde cae 치guila en el $i$-칠simo resultado. As칤, tenemos

$$A = A_{1} \cap A_{2} \cap A_{3}$$

Vamos a crear a nuestro primer evento $A_{1}$, es decir donde en la primera coordenada es 치guila:

```python
A_1 = {om for om in Omega if om[0]=="A" } #la primer coordenada
A_1
```

```python
A_2 = {om for om in Omega if om[1]=="A" } #la segunda coordenada es 치guila
A_2
```

```python
A_3 = {om for om in Omega if om[2]=="A" } #la tercer coordenada es 치guila
A_3
```

Para intersectar los conjuntos usamos la funci칩n `intersection()`

```python
A = A_1.intersection(A_2.intersection(A_3))
A
```

Vamos a calcular la probabilidad del evento A

```python
#P_A denota a la probabilidad (P) del evento A
P_A = len(A)/len(Omega)
P_A
```

En vez del d칠cimal, podemos tener el n칰mero expresado eb fracci칩n, lo cual se hace como sigue:

Vamos a crear una funci칩n que calcule la probabilidad de un evento $E$ dentro de un espacio muestal $\Omega$:

```python
from fractions import Fraction

def p(E, Omega): #E un evento
  p = Fraction(len(E),len(Omega))
  return p
```

# Ejercicio 1: Todos los resultados son soles

## Desarrollo

Sea $A$ = $\{  (x_1, x_2, x_3) : x_i = A \vee x_i = S \}$ ;

Para  $i = \{1,2,3\}$

$$\mathbb{P}[\text{Todos Sol}] = \frac{\#\{A \in \Omega : x_i = S; \forall x_i\in A \}}{\#\Omega} =  \frac{\#\{(S, S, S)\}}{\#\Omega} = \frac{1}{8} = \mathbb{P}[(S,S,S)]$$


$$$$

## Implementaci칩n

```python
# Espacio Muestral
Omega = set(product({"A","S"},repeat=3))

# Lanzamientos con Soles
S_1 = {om for om in Omega if om[0] == "S"} # Sol en la primer coordenada
S_2 = {om for om in Omega if om[1] == "S"} # Sol en la segunda coordenada
S_3 = {om for om in Omega if om[2] == "S"} # Sol en la tercer coordenada
S = S_1.intersection(S_2.intersection(S_3)) # Intersecci칩n

# Funci칩n que calcule la probabilidad de un evento 洧냦 dentro de un espacio muestal 풜
def p(E, Omega):
  p = Fraction(len(E),len(Omega))
  return p

# Probabilidad de que todos los lanzamientos sean Sol
p(S, Omega)
```
